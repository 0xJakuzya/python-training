## введение в множества

множества мы используем если хотим существует ли что-то или нет, а другая информация нам не нужна. это как сумка с ключами, если мы хотим добавить информацию о ключах, то используем словари

```python
empty_set = {}
even_numbers = {0, 2, 4, 8, 10}
odd_numbers = {1, 3, 5, 7, 9}

set('letters') ==> {'l', 't', 'e', 'r', 's'} # строка --> множество букв строки, не содержит дубликатов строки!!
set(['dascher', 'dancer', 'pancer', 'mason-dixsen']) ==> {'dascher', 'dancer', 'pancer', 'mason-dixsen'} # список --> множество
set(('a', 'b', 'c')) ==> {'a', 'b', 'c'} # кортеж --> множество
```
порядок ключей в множестве не имеет значения

операция над множествами:

- `объединение` - множество, состоящее из элементов, принадлежащих хотя бы одному из объединяемых множеств. используется символ U
- `пересечение` - множество, состоящее из элементов, принадлежащих одновременно каждому из пересекающихся множеств
- `разность` - множество, в которое входят только элементы первого множества, не входящие во второе множество
- `симметрическая разность` - множеств – множество, включающее все элементы исходных множеств, не принадлежащие одновременно обоим исходным множествам
- `дополнение` - множество всех элементов, в нем не содержащихся

## книги на прочтение

Ученики 10-го класса онлайн-школы получили задание прочесть на летних каникулах три книги:

- "Что такое математика?";
- "Математическая составляющая";
- "100 гениальных идей по математике".
Оказалось, что

- `n` - прочитали первую книгу 
- `m` – прочитали вторую книгу 
- `k` – прочитали третью книгу
- `x` - прочитали первую или вторую, или обе эти книги
- y - прочитали вторую или третью, или обе
- z учеников – первую и третью, или хотя бы одну из этих двух книг.
- t учеников - прочитали все книги
- a - всего учеников

Напишите программу, которая выводит, сколько учеников:

- прочитали только одну книгу
- прочитали только две книги
- не прочитали ни одной из рекомендованных книг

````python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

books = ['что такое математика?', 
         'математическая составляющая', 
         '100 гениальных идей по математике']

def read_book(n, m, k, x, y, z, t, a):

    nm = n + m - x 
    mk = m + k - y
    nk = n + k - z

    read_nm_book = nm - t 
    read_mk_book = mk - t
    read_nk_book = nk - t

    read_only_n = n - (read_nm_book + read_nk_book + t)
    read_only_m = m - (read_nm_book + read_mk_book + t)
    read_only_k = k - (read_mk_book + read_nk_book + t)

    total_readers = (read_only_n + read_only_m + read_only_k + 
                     read_nm_book + read_mk_book + read_nk_book + t)
    
    print(f'прочитали одну книгу: {read_only_n + read_only_m + read_only_k}')
    print(f'прочитали две книги: {read_nm_book + read_mk_book + read_nk_book}')
    print(f'не читали ни одной: {a-total_readers}')
    
    return (read_only_n, read_only_m, read_nm_book, 
            read_only_k, read_nk_book, read_mk_book, t)

n, m, k, x, y, z, t, a = (int(input()) for i in range(8))
print()
subsets = read_book(n, m, k, x, y, z, t, a)

plt.figure(figsize=(10, 8))
venn3(subsets=subsets, set_labels=books)
plt.show()
````
