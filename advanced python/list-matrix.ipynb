### Треугольник Паскаля 1 

````python
def pascal(n):
    row = [1]
    for i in range(1, n+1):
        value = row[i-1] * (n-i+1)//i
        row.append(value)
    return row

n = int(input())
print(pascal(n))
````

### Треугольник Паскаля 2

````python
def pascal(n):
    row = [1]
    for i in range(1, n+1):
        value = row[i-1] * (n - i + 1) // i
        row.append(value)
    return row

n = int(input())
for i in range(n):
    print(' '.join(map(str, pascal(i))))
````

### Упаковка дубликатов
```
Sample Input 1:

a b c d

Sample Output 1:

[['a'], ['b'], ['c'], ['d']]
```

````python
def pack_duplicated(s):
    lst_duplicated = [[s[0]]]
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            lst_duplicated[-1].append(s[i])
        else:
            lst_duplicated.append([s[i]])
    return lst_duplicated

s = input().split()
print(pack_duplicated(s))
````

### Вывести матрицу 1

````python
n = int(input())
m = int(input())
matrix = [[0]*m for _ in range(n)]
for r in range(n):
    for c in range(m):
        matrix[r][c] = input()
        print(*matrix[r])
````

### Вывести матрицу 2

````python
n = int(input())
m = int(input())
matrix = [[0]*m for _ in range(n)]
for i in range(n):
    for j in range(m):
        matrix[i][j] = input()
    print(*matrix[i])
print()
for j in range(m):
    for i in range(n):
        print(matrix[i][j], end=' ')
    print()
````

### След матрицы

````python
n = int(input())
matrix = []
s = 0
for i in range(n):
    temp = [int(num) for num in input().split()]
    matrix.append(temp)
    s += matrix[i][i]
print(s)
````

### Больше среднего

````python
n = int(input())
matrix = []

for i in range(n):
    temp = [int(num) for num in input().split()]
    matrix.append(temp)
    s = 0
    sp = sum(matrix[i])/n
    for j in range(n):
        if matrix[i][j]>sp:
            s+=1
    print(s)
````

### Максимальный в области 1

````python
n = int(input())
matrix = []
for i in range(n):
    temp = [int(num) for num in input().split()]
    matrix.append(temp)
maximum = matrix[0][0]
for i in range(n):
    for j in range(i+1):
        if matrix[i][j] > maximum:
                maximum = matrix[i][j]
print(maximum)
````

### Максимальный в области 2

````python
n = int(input())
matrix = []
for i in range(n):
    temp = [int(num) for num in input().split()]
    matrix.append(temp)
maximum = matrix[0][0]
for i in range(n):
    for j in range(n):
        if ((i >= j and i <= n - 1 - j) or (i <= j and i >= n - 1 - j)) and matrix[i][j] > maximum:
            maximum = matrix[i][j]
````
