```python
def two_sum(n, target):
        values = {}
        for i, v in enumerate(n):
            a = target - v
            if a in values:
                return [values[a], i]
            values[v] = i
        return []

numbers_list = [2, 7, 11, 15]
target_sum = 9
print(two_sum(numbers_list, target_sum))
```

```python
def merge_alternately(word1, word2):
    result = ""
    for char1, char2 in zip(word1, word2):
        result += char1 + char2
    result += word1[len(word2):] + word2[len(word1):]
    return result
    
word1 = "abc"
word2 = "pqr"
print(merge_alternately(word1, word2))
```
